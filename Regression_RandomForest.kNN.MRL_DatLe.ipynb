```python
# One-hot encode the data using pandas get_dummies
data = pd.get_dummies(data)
# Display the first 5 rows of the last 4226 columns
data.iloc[:,5:].head(5)

data_np=data.to_numpy()
print(data_np.shape)
# output
Age_np=data_np[:,139]  
# The rest of features are the input
input_data=data_np[:,0:139]    
print(Age_np.shape)
print(input_data.shape)

print(Age_np)
```

```python
# Using Skicit-learn to split data into training and testing sets (80% and 20%)
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets - x is features and y is labels
train_features, test_features, train_labels, test_labels = train_test_split(input_data, Age_np, test_size = 0.2, random_state = 42)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
# Import the model we are using: Random Forest
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
# Train the model on training data
rf.fit(train_features, train_labels);
# Use the forest's predict method on the test and train data
predictions = rf.predict(test_features)

# Calculate the absolute errors
errors = abs(predictions - test_labels)

# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
```

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / test_labels)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```

```python
# Import the model we are using: knn model
from sklearn.neighbors import KNeighborsRegressor
knn_model = KNeighborsRegressor(n_neighbors=3)

# Apply the model to training dataset
knn_model.fit(train_features,train_labels)
predictions = knn_model.predict(test_features)

# Calculate the absolute errors
errors = abs(predictions - test_labels)

# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
```

```python
#Calculate and display accuracy
names = []
scores = []
knn_model.fit(train_features,train_labels)
names.append("kNN")
knn_accuracy = knn_model.score(train_features, np.ravel(train_labels))*100
scores.append(knn_accuracy)
print("knn model accuracy: ", knn_accuracy)

y_pred_knn_train = knn_model.predict(train_features)
sns.scatterplot(train_labels,y_pred_knn_train)
```

```python
from sklearn.linear_model import LinearRegression

multiple_linear_reg = LinearRegression(fit_intercept=False)
multiple_linear_reg.fit(train_features, train_labels)
```

```python
from sklearn.model_selection import cross_val_predict  # For K-Fold Cross Validation
from sklearn.metrics import r2_score  # For find accuracy with R2 Score
from sklearn.metrics import mean_squared_error  # For MSE
from math import sqrt  # For squareroot operation

#Prediction of training and testing data
y_pred_MLR_train = multiple_linear_reg.predict(train_features)
y_pred_MLR_test = multiple_linear_reg.predict(test_features)

accuracy_MLR_train = r2_score(train_labels, y_pred_MLR_train)
print("Training Accuracy for Multiple Linear Regression Model: ", accuracy_MLR_train*100)

accuracy_MLR_test = r2_score(test_labels, y_pred_MLR_test)
print("Testing Accuracy for Multiple Linear Regression Model: ", accuracy_MLR_test*100)

RMSE_MLR_train = sqrt(mean_squared_error(train_labels, y_pred_MLR_train))
print("RMSE for Training Data: ", RMSE_MLR_train)

RMSE_MLR_test = sqrt(mean_squared_error(test_labels, y_pred_MLR_test))
print("RMSE for Testing Data: ", RMSE_MLR_test)

```
