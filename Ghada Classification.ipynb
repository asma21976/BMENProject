```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler 
from sklearn import preprocessing 
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import metrics 

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score
```

```python
dataset= pd.read_csv('/Users/ghadaeldib/Desktop/breast-cancer.csv')
dataset.head()
```

```python
Y=['diagnosis']
X=list(set(list(dataset.columns))-set(Y))
dataset[X]=dataset[X]/dataset[X].max()
dataset.describe().transpose()
```

```python
#splitting dataset into training and testing dataset
from sklearn.model_selection import train_test_split
X=dataset[X].values
Y=dataset[Y].values

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)#test size is 20% of the data, remaining 80 is training
print(X_train.shape, X_test.shape)
```

```python
#feature scaling to convert all variables to the same scale
scaler =StandardScaler()
scaler

X_train = scaler.fit_transform(X_train) #use fit transform on training set
X_test = scaler.transform(X_test) #use transform on test set so the same mean and variance 
                                  #used on the training data as is used to transform our test data
```

```python
X_train.mean(axis=0) #standardized data should have mean of 0
```

```python
X_train.std(axis=0) #standardized data should have a unit variance
```

# PLS Discriminant Analysis

```python
#for PLS Discriminate Analysis
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
clf = LinearDiscriminantAnalysis()
clf.fit(X_train, np.ravel(Y_train))

clf.predict(X_train)
Y_predict_test=clf.predict(X_test)

print(classification_report(Y_test, Y_predict_test)) 
print(accuracy_score(Y_test, Y_predict_test))

cf=confusion_matrix(Y_test, Y_predict_test)

ax = sns.heatmap(cf, annot=True, cmap='Blues')

ax.set_title('Seaborn Confusion Matrix with labels\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['False','True'])
ax.yaxis.set_ticklabels(['False','True'])

## Display the visualization of the Confusion Matrix.
plt.show()
```

# Bagged Tree

```python
#Bagged Tree
from sklearn.ensemble import BaggingClassifier
```

```python
clf = BaggingClassifier()
clf.fit(X_train, np.ravel(Y_train))

clf.predict(X_train)
Y_predict_test=clf.predict(X_test)

print(classification_report(Y_test, Y_predict_test)) 
print(accuracy_score(Y_test, Y_predict_test))

cf=confusion_matrix(Y_test, Y_predict_test)

ax = sns.heatmap(cf, annot=True, cmap='Blues')

ax.set_title('Seaborn Confusion Matrix with labels\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['False','True'])
ax.yaxis.set_ticklabels(['False','True'])

## Display the visualization of the Confusion Matrix.
plt.show()
```

# Boosted Tree

```python
#Boosted Tree
from sklearn.ensemble import GradientBoostingClassifier
```

```python
clf = GradientBoostingClassifier()
clf.fit(X_train, np.ravel(Y_train))

clf.predict(X_train)
Y_predict_test=clf.predict(X_test)

print(classification_report(Y_test, Y_predict_test)) 
print(accuracy_score(Y_test, Y_predict_test))

cf=confusion_matrix(Y_test, Y_predict_test)

ax = sns.heatmap(cf, annot=True, cmap='Blues')

ax.set_title('Seaborn Confusion Matrix with labels\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['False','True'])
ax.yaxis.set_ticklabels(['False','True'])

## Display the visualization of the Confusion Matrix.
plt.show()
```
