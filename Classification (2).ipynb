```python
#Loading the data
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('C:/Users/hashm/OneDrive/Desktop/breast-cancer.csv')
```

```python
dataset.head()
```

```python
Y=['diagnosis']
X=list(set(list(dataset.columns))-set(Y))
dataset[X]=dataset[X]/dataset[X].max()
dataset.describe().transpose()
```

```python
#splitting dataset into training and testing dataset
from sklearn.model_selection import train_test_split
X=dataset[X].values
Y=dataset[Y].values

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)#test size is 20% of the data, remaining 80 is training
print(X_train.shape, X_test.shape)
```

```python
#feature scaling to convert all variables to the same scale
scaler =StandardScaler()
scaler

X_train = scaler.fit_transform(X_train) #use fit transform on training set
X_test = scaler.transform(X_test) #use transform on test set so the same mean and variance 
                                  #used on the training data as is used to transform
```

```python
X_train.mean(axis=0) #standardized data should have mean of 0
```

```python
X_train.std(axis=0) #standardized data should have a unit variance
```

```python
#Neural Network
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
import seaborn as sns

clf = MLPClassifier(random_state=0)
clf.fit(X_train, np.ravel(Y_train))

clf.predict(X_train)
Y_predict_test=clf.predict(X_test)

print(classification_report(Y_test, Y_predict_test)) 
print(accuracy_score(Y_test, Y_predict_test))

cf=confusion_matrix(Y_test, Y_predict_test)

ax = sns.heatmap(cf, annot=True, cmap='Blues')

ax.set_title('Neural Network Confusion Matrix\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['False','True'])
ax.yaxis.set_ticklabels(['False','True'])

## Display the visualization of the Confusion Matrix.
plt.show()
```
