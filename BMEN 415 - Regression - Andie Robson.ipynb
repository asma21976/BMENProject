```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler

from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor

from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel ("Volumetric_features.xlsx")
```

```python
data.head()
```

```python
#matrix of features
X = data.iloc[:,1:138].values
#vectors of age  
Y = data.iloc[:,139].values
```

```python
#splitting dataset into training and testing dataset
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0) #test size is 20% of the data, remaining 80 is training
```

```python
#feature scaling to convert all variables to the same scale
scaler =StandardScaler()
scaler
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_train.mean(axis=0) #standardized data should have mean of 0
```

```python
X_test = scaler.transform(X_test) #use transform on test set so the same mean and variance 
                                  #used on the training data as is used to transform our test data
```

```python
#Basic Tree Regression Model
regr = DecisionTreeRegressor()
regr.fit(X_train, Y_train)

regr.predict(X_train)
Y_predict_test=regr.predict(X_test)

#Performance metrics
VAR = metrics.explained_variance_score(Y_test, Y_predict_test)
MSE = metrics.mean_squared_error(Y_test, Y_predict_test)
R2=metrics.r2_score(Y_test, Y_predict_test)

print('Explained Variance:', VAR)
print('Mean Square Error:', MSE)
print('R-Squared', R2)

#Density scatter plot of true brain age versus predicted brain age from test set data
from scipy.stats import gaussian_kde

x = Y_test
y = Y_predict_test

# Calculate point density
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)

# Sorting points based on density
sort = z.argsort()
x, y, z = x[sort], y[sort], z[sort]

fig, ax = plt.subplots()
ax.scatter(x, y, c=z, s=50)
```

```python
# MLR Regression Model
mlr = LinearRegression()  
mlr.fit(X_train, Y_train)

mlr.predict(X_train)
Y_predict_test=mlr.predict(X_test)

#Performance metrics
VAR = metrics.explained_variance_score(Y_test, Y_predict_test)
MSE = metrics.mean_squared_error(Y_test, Y_predict_test)
R2=metrics.r2_score(Y_test, Y_predict_test)

print('Explained Variance:', VAR)
print('Mean Square Error:', MSE)
print('R-Squared', R2)

#Density scatter plot of true brain age versus predicted brain age from test set data
from scipy.stats import gaussian_kde

x = Y_test
y = Y_predict_test

# Calculate point density
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)

# Sorting points based on density
sort = z.argsort()
x, y, z = x[sort], y[sort], z[sort]

fig, ax = plt.subplots()
ax.scatter(x, y, c=z, s=50)
```

```python
# Neural Network Regression Model
NNreg = MLPRegressor(solver='adam', alpha=1e-5,  max_iter=5000, random_state=1) #, hidden_layer_sizes=(5, 2)
NNreg.fit(X_train, Y_train)

NNreg.predict(X_train)
Y_predict_test=NNreg.predict(X_test)

#Performance metrics
VAR = metrics.explained_variance_score(Y_test, Y_predict_test)
MSE = metrics.mean_squared_error(Y_test, Y_predict_test)
R2=metrics.r2_score(Y_test, Y_predict_test)

print('Explained Variance:', VAR)
print('Mean Square Error:', MSE)
print('R-Squared', R2)

#Density scatter plot of true brain age versus predicted brain age from test set data
from scipy.stats import gaussian_kde

x = Y_test
y = Y_predict_test

# Calculate point density
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)

# Sorting points based on density
sort = z.argsort()
x, y, z = x[sort], y[sort], z[sort]

fig, ax = plt.subplots()
ax.scatter(x, y, c=z, s=50)
```
