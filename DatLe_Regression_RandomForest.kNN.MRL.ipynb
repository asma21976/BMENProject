```python
# Select the input brain volume vector (X) and the output is Age  (y)
X = data.drop(columns=['S.No', 'Age', 'dataset'])
Y = data['Age']
```

```python
print('The shape of our data is:', data.shape)
```

```python
# Using Skicit-learn to split data into training and testing sets (80% and 20%)
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets - x is features and y is labels
train_features, test_features, train_labels, test_labels = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
#feature scaling to convert all variables to the same scale
from sklearn.preprocessing import StandardScaler
scaler =StandardScaler()
scaler
```

```python
train_features = scaler.fit_transform(train_features) #use fit transform on training set
test_features = scaler.transform(test_features) #use transform on test set so the same mean and variance 
```

```python
train_features.mean(axis=0) #standardized data should have mean of 0
```

```python
train_features.std(axis=0) #standardized data should have a unit variance
```

```python
# Import the model we are using: Random Forest
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
# Train the model on training data
rf.fit(train_features, train_labels);
# Use the forest's predict method on the test data
prediction0 = rf.predict(test_features)

# compare the predictions to the known values (y_test)
print('r square value:', metrics.r2_score(test_labels, prediction0))
print('mean absolute error (years):', metrics.mean_absolute_error(test_labels, prediction0))
print('mean square error:', metrics.mean_squared_error(test_labels, prediction0))
print('root mean sqaure error (years):', np.sqrt(metrics.mean_squared_error(test_labels, prediction0)))
print('Explained Variance:',metrics.explained_variance_score(test_labels, prediction0))
```

```python
y_pred_rf_train = rf.predict(train_features)
y_pred_rf_test = rf.predict(test_features)
```

```python
sns.scatterplot(train_labels,y_pred_rf_train)
```

```python
#Density scatter plot of true brain age versus predicted brain age from testing set data
from scipy.stats import gaussian_kde

x = test_labels
y = y_pred_rf_test

# Calculate point density
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)

fig, ax = plt.subplots()
ax.scatter(x, y, c=z, s=50)
```

```python
# Import the model we are using: knn model
from sklearn.neighbors import KNeighborsRegressor
knn_model = KNeighborsRegressor(n_neighbors=3)

# Apply the model to training dataset
knn_model.fit(train_features,train_labels)
prediction1 = knn_model.predict(test_features)

# compare the predictions to the known values (y_test)
print('r square value:', metrics.r2_score(test_labels, prediction1))
print('mean absolute error (years):', metrics.mean_absolute_error(test_labels, prediction1))
print('mean square error:', metrics.mean_squared_error(test_labels, prediction1))
print('root mean sqaure error (years):', np.sqrt(metrics.mean_squared_error(test_labels, prediction1)))
print('Explained Variance:',metrics.explained_variance_score(test_labels, prediction1))
```

```python
y_pred_knn_train = knn_model.predict(train_features)
sns.scatterplot(train_labels,y_pred_knn_train)
```

```python
#Density scatter plot of true brain age versus predicted brain age from training set data
from scipy.stats import gaussian_kde
y_pred_knn_test = knn_model.predict(test_features)
x = test_labels
y = y_pred_knn_test

# Calculate point density
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)

fig, ax = plt.subplots()
ax.scatter(x, y, c=z, s=50)
```

```python
#Importing model: MLR
from sklearn.linear_model import LinearRegression
multiple_linear_reg = LinearRegression(fit_intercept=False)
multiple_linear_reg.fit(train_features, train_labels)
from sklearn.metrics import r2_score  # For find accuracy with R2 Score
from sklearn.metrics import mean_squared_error  # For MSE
from math import sqrt  # For squareroot operation

#Prediction of training and testing data
y_pred_KCV_train =multiple_linear_reg.predict(train_features)
y_pred_KCV_test =multiple_linear_reg.predict(test_features)

prediction2 = multiple_linear_reg.predict(test_features)
print('r square value:', metrics.r2_score(test_labels, prediction2))
print('mean absolute error (years):', metrics.mean_absolute_error(test_labels, prediction2))
print('mean square error:', metrics.mean_squared_error(test_labels, prediction2))
print('root mean sqaure error (years):', np.sqrt(metrics.mean_squared_error(test_labels, prediction2)))
print('Explained Variance:',metrics.explained_variance_score(test_labels, prediction2))
```
